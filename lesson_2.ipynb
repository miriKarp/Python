```python
# years = range(2010, 2021)
# consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
# consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
# consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
years = [2010, 2011, 2012, 2013, 2014]
coal_consumption = [100, 120, 110, 105, 95]
gas_consumption = [80, 85, 90, 88, 82]
solar_consumption = [5, 8, 12, 15, 20]
fig, axs = plt.subplots(2, sharex=True, figsize=(10, 8))
axs[0].plot(years, coal_consumption, label='Coal', marker='o')
axs[0].plot(years, gas_consumption, label='Gas', marker='s')
axs[0].plot(years, solar_consumption, label='Solar', marker='^')
axs[0].set_title('Energy Consumption by Years')
axs[0].set_ylabel('Energy Consumption')
axs[0].legend()

# Plotting energy consumption by sources
axs[1].bar(years, coal_consumption, label='Coal', alpha=0.7)
axs[1].bar(years, gas_consumption, label='Gas', alpha=0.7)
axs[1].bar(years, solar_consumption, label='Solar', alpha=0.7)
axs[1].set_title('Energy Consumption by Sources')
axs[1].set_xlabel('Years')
axs[1].set_ylabel('Energy Consumption')
axs[1].legend()

plt.tight_layout()
plt.show()
```
